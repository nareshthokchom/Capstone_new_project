```python
#importing all the neccesary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the dataset
df = pd.read_csv('Dentistry Dataset.csv')
df.head()
```

```python
#checking the shape of the data
df.shape
```

```python
#checking null values
df.isnull().sum()
```

```python
#dropping unnessesay columns
df = df.drop(df[['Sample ID','Sl No']],axis=1)
df.head()
```

```python
#encoding the categorical values
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df["Gender"] = encoder.fit_transform(df['Gender'])
df.head()
```

```python
#keeping the gender in the last column
temp = df.pop("Gender")
df["Gender"] = temp
df.head()
```

```python
#checking class imbalance
df["Gender"].value_counts()
```

```python
#checking outliers
for col in df.columns:
    mean = df[col].mean()
    std_dev = df[col].std()
    outliers = df[(df[col] < mean - 3 * std_dev) | (df[col] > mean + 3 * std_dev)]
    print(f"Outliers in {col}: {len(outliers)}")
```

```python
# Visualize data with outliers using a box plot
plt.boxplot(df)
plt.title('Box Plot with Outliers')
plt.show()
```

```python
#dependent variable and independent variable
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
#Check for Multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data['Feature'] = x.columns
vif_data['VIF'] = [variance_inflation_factor(x, i) for i in range(x.shape[1])]

# Display features with high VIF
print(vif_data)
```

```python
#dropping high collinear columns
df = df.drop(columns=["right canine width casts","right canine width intraoral","left canine width intraoral"],axis=1)
df.head()
```

```python
#plotting collinear heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),annot=True,cmap="coolwarm",vmin = -1,vmax = 1)
```

```python
#dependent and independent variable
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
#normalizing again
from sklearn.preprocessing import MinMaxScaler
norm = MinMaxScaler()
x_normalized = norm.fit_transform(x)
```

```python
#splitting the data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_normalized,y,test_size=0.2,random_state=1)
```

```python
#importing all the models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
lr_model = LogisticRegression()
dc_model = DecisionTreeClassifier()
rf_model = RandomForestClassifier()
xg_model = XGBClassifier()
```

```python
#fitting all the models
models =  [lr_model,dc_model,rf_model,xg_model]
#fit all the models
for i in models:
    i.fit(x_train,y_train)
```

```python
#evaluating the models
from sklearn.metrics import roc_curve,auc,RocCurveDisplay,confusion_matrix
for i in models:
    #predict the values
    preds = i.predict(x_test)
    # Get predicted probabilities for the positive class (1)
    y_pred_prob = i.predict_proba(x_test)[:, 1]
    # Calculate the False Positive Rate (FPR) and True Positive Rate (TPR)
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
    # Calculate the AUC score
    roc_auc = auc(fpr, tpr)
    #Using RocCurveDisplay from scikit-learn
    roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr).plot()
    plt.title('ROC Curve')
    plt.show()
    print(f'AUC score:{roc_auc:.2f}')
    print(confusion_matrix(preds,y_test))
    print(f"for the model {i}")
```

Here the random forest and the xgboost model has an AUC score of 97% making it the best model among all the models
